# Digital Signal Processing (DSP)

An infinite impulse response (IIR) filter is a type of digital filter where the output of the filter depends on all past inputs. The benifit of IIR DSP is you can can easily create and modify the filter in real time.

Here is an example of an IIR filter written in Python that can change based on a desired cutoff frequency:

## Digital biquad filter

In signal processing, a digital biquad filter is a second order recursive linear filter, containing two poles and two zeros. 

### Difference equation

```
y[n] = b0.x[n] + b1.x[n-1] + b2.x[n-2] + a1.y[n-1] + a2.y[n-2]
```

### Quadratic function

```
H(z) =  b0 + b1.[z-1] + b2.[z-2]
       -------------------------
        1 + a1.[z-1] + a2.[z-2]
```

A value that causes the numerator to be zero is a transfer-function zero, and a value that causes the denominator to be zero is a transfer-function pole.

The b0, b1 and b2 coefficients determine zeros, and a1, a2 determine the position of the poles.

```
             (zeros)       (poles) 
x[n]------+---->b0-->( + )---------+-----> y[n]
          |           /|\          |
         \|/           |          \|/
       [ z-1 ]         |        [ z-1 ]
          |            |           |
          +---->b1-->( + )<- -a1<--+
          |           /|\          |
         \|/           |          \|/
       [ z-1 ]         |        [ z-1 ]
          |            |           |
          +---->b2-->( + )<- -a2<--+
```

````python
import math

class IIRFilter:
    
    y_1 = 0.0;
    y_2 = 0.0;
    x_1 = 0.0;
    x_2 = 0.0;
    
    def __init__(self, sample_rate, cutoff_frequency):
        self.sample_rate = sample_rate
        self.cutoff_frequency = cutoff_frequency;
        omega = 2 * math.pi * self.cutoff_frequency / self.sample_rate
        alpha = math.sin(omega) / (2 * math.sqrt(2))
        a_0 = 1 + alpha
        self.a_1 = -2 * math.cos(omega)
        self.a_2 = 1 - alpha
        self.b_0 = (1 - math.cos(omega)) / 2
        self.b_1 = 1 - math.cos(omega)
        self.b_2 = (1 - math.cos(omega)) / 2
        #normalise
        self.b_0 = self.b_0 / a_0;
        self.b_1 = self.b_1 / a_0
        self.b_2 = self.b_2 / a_0;
        self.a_1 = self.a_1 / a_0;
        self.a_2 = self.a_2 / a_0;
        
    def process(self, x):
        # Get latest value
        y = self.b_0 * x + self.b_1 * self.x_1 + self.b_2 * self.x_2 - self.a_1 * self.y_1 - self.a_2 * self.y_2

        self.x_2 = self.x_1
        self.x_1 = x
        self.y_2 = self.y_1
        self.y_1 = y

        return y

````

## Steady State Detection (SSD)

To know if a signal is viable and not noise, you may want to know if the signal is "steady". A simple method of detecting a steady state is to look at the signal deviation within a time window.

````python
def SSD( data , ptr, past = 10 ):
    if ptr < past:
        return False 
    v1 = data[ptr]
    deltas = 0;
    for i in range(1,past):
        v2= data[ ptr - i ];
        deltas += abs( v1 - v2 );
    return deltas < 0.025;
````

## Peak Finding

You often will want to know when a signal trend is at a point of inflection. This can be done why looking at when the vector sign of the filtered signal changes.

````python
def peakFinding( data , ptr ):
    if ptr < 6:
        return False
    if not hasattr( peakFinding, "lastVector"):
        peakFinding.lastVector = 0
    v = 0;
    for i in range(0, 6, 2):
        current = data[ptr - i];
        previous = data[ptr - i - 1];
        v += ( current - previous );
    v /= 4;
    
    if peakFinding.lastVector > 0.001 and v < 0.001:
        peakFinding.lastVector = v;
        return True; # Peak inflection
    peakFinding.lastVector = v;
    return False;
````

````python
%matplotlib inline
import random
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

#Settings:
SAMPLE_RATE = 10; 
FREQUENCY = 0.1;
end = 300;

random.seed(5)
DSP = IIRFilter( SAMPLE_RATE , FREQUENCY );

# create empty lists for the x and y data
x = [ ];
raw = [];
fltr = [];
steady = ([],[]);
peak = ([],[]);

# create the figure and axes objects
fig = plt.figure()
ax = plt.axes( xlim=(0,500),ylim=(0,1) );

lineRaw = plt.plot([], [])[0];
lineFltr = plt.plot([], [])[0];
pltSteady = plt.plot([], [],'o', color="red")[0];
pltPeak = plt.plot([], [],'x', color="yellow")[0];

# function that draws each frame of the animation
def animate(i):
    v = random.random();
    
    ptr = len(x) - 1;
    
    if SSD( fltr , ptr ):
        steady[0].append( ptr )
        steady[1].append( fltr[-1] )
        
    if peakFinding( fltr , ptr ):
        peak[0].append( ptr )
        peak[1].append( fltr[-1] )
    
    raw.append( v ) # grab a random integer to be the next y-value in the animation
    fltr.append( DSP.process( v ) ) # grab a random integer to be the next y-value in the animation
    x.append(i)
    
    lineRaw.set_data( x,raw)
    lineFltr.set_data( x,fltr)
    
    pltSteady.set_data( steady[0], steady[1] )
    pltPeak.set_data( peak[0], peak[1] )
    
    return lineRaw, lineFltr, pltSteady, pltPeak

anim = FuncAnimation(fig, animate, frames=500, interval=50, repeat=False)
HTML(anim.to_html5_video())
````
